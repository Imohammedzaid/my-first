```
import random
import time

# Define workout categories
warmup = ["Jumping Jacks", "High Knees", "Arm Circles", "Butt Kicks"]
strength = ["Push-ups", "Squats", "Lunges", "Plank", "Burpees"]
cardio = ["Mountain Climbers", "Skipping Rope", "Sprint in Place", "Jump Squats"]
cooldown = ["Stretch Hamstrings", "Child's Pose", "Cat-Cow Stretch", "Shoulder Stretch"]

# Function to pick random exercises
def pick_exercises(category, n=2):
    return random.sample(category, n)

# Generate a workout plan
print("Your Workout Plan\n")

print("Warm-up:", pick_exercises(warmup))
print("Strength:", pick_exercises(strength))
print("Cardio:", pick_exercises(cardio))
print("Cooldown:", pick_exercises(cooldown))

# Optional: Timer for each exercise
def exercise_timer(seconds=30):
    for i in range(seconds, 0, -1):
        print(f"{i} seconds left...", end="\r")
        time.sleep(1)
    print("Done!\n")

# Example: Run a 30-second timer for Push-ups
print("\nStart: Push-ups")
exercise_timer(30)
```
