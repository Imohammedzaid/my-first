```
import random

# Ensure colorama is available (installs automatically if missing)
try:
    from colorama import Fore, Style, init
except ImportError:
    import sys, subprocess
    subprocess.check_call([sys.executable, "-m", "pip", "install", "colorama"])
    from colorama import Fore, Style, init

# Initialize colorama
init(autoreset=True)

# Define celebrity categories
actors = ["Leonardo DiCaprio", "Emma Watson", "Denzel Washington", "Scarlett Johansson"]
singers = ["Taylor Swift", "Ed Sheeran", "Beyonc√©", "Adele"]
athletes = ["Cristiano Ronaldo", "Serena Williams", "Virat Kohli", "Usain Bolt"]
influencers = ["MrBeast", "Charli D'Amelio", "Khaby Lame", "PewDiePie"]

def pick_items(category, n=2):
    n = max(1, min(n, len(category)))
    return random.sample(category, n)

print(Fore.CYAN + "Celebrity Lineup\n" + Style.RESET_ALL)
print(Fore.YELLOW + "Actors:" + Style.RESET_ALL, pick_items(actors, 2))
print(Fore.RED + "Singers:" + Style.RESET_ALL, pick_items(singers, 2))
print(Fore.GREEN + "Athletes:" + Style.RESET_ALL, pick_items(athletes, 2))
print(Fore.BLUE + "Influencers:" + Style.RESET_ALL, pick_items(influencers, 2))
```
